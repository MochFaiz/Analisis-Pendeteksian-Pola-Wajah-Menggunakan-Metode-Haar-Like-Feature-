```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
face_cascade_file = 'Cascade Classifier/face-detect.xml'
face_cascade = cv2.CascadeClassifier(face_cascade_file)
```

```python
image_path = 'fotokamu.jpg'
img = cv2.imread(image_path)                                       
img = cv2.resize(img, (640, 360), interpolation=cv2.INTER_LINEAR)

rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)                  
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)                

faces = face_cascade.detectMultiScale(gray_img, 1.3, 1)     
                                                                
for x, y, w, h in faces:                                           
    cv2.rectangle(rgb_img, (x, y), (x+w, y+h), (0, 255, 0), 3) 


plt.imshow(rgb_img)
plt.show()
```
